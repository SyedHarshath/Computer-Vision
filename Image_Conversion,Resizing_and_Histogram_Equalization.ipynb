### IMAGE 1

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
image = cv2.imread("lenna.png")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Lenna Image")
plt.show()
```

### RGB TO GRAYSCALE CONVERSION USING INBUILT FUNCTION

```python
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
cv2.imshow("GRAY IMAGE",image_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### IMAGE RESIZING

```python
resized_image = cv2.resize(image,(100,100))
cv2.imshow("RESIZED IMAGE",resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
print("Size of the Resized Image : ",resized_image.size)
```

```python
print("Dimensions of Original Image")
h , w = image.shape[:2]
print("Height : ",h)
print("Width : ",w)
h , w = resized_image.shape[:2]
print("Dimensions of Resized image ")
print("Height : ",h)
print("Width : ",w)
```

### HISTROGRAM EQUALIZATION USING INBUILT FUNCTION

```python
hist = cv2.calcHist([image_gray],[0],None,[256],[0,256])
plt.plot(hist,color='maroon')
plt.title("Histogram Equalization")
plt.xlabel("Pixel")
plt.ylabel("Frequency")
plt.show()
```

### MANUAL RGB TO GRAY SCALING

```python
gamma = 1.04
r_const,g_const,b_const = 0.2126,0.7152,0.0722
R,G,B= image_rgb[:,:,0],image_rgb[:,:,1],image_rgb[:,:,2]
greyscale_image = r_const*R**gamma + g_const*G**gamma + b_const*B**gamma
greyscale_image
```

```python
fig = plt.figure(1)
img1, img2 = fig.add_subplot(121), fig.add_subplot(122)
img1.imshow(image_rgb)
img1.axis('off')
img2.imshow(greyscale_image, cmap='gray')
img2.axis('off')
plt.show()
```

### MANUAL HISTOGRAM EQUALIZATION

```python
greyscale_image = greyscale_image.astype(np.uint8)
hist = np.zeros(256)
rows, cols = greyscale_image.shape

for i in range(rows):
    for j in range(cols):
        hist[greyscale_image[i, j]] += 1

cdf = np.cumsum(hist)
cdf_min = cdf.min()
cdf_max = cdf.max()
cdf_normalized = ((cdf - cdf_min) / (cdf_max - cdf_min)) * 255
cdf_normalized = cdf_normalized.astype('uint8')

equalized_image = np.zeros_like(greyscale_image)

for i in range(rows):
    for j in range(cols):
        equalized_image[i, j] = cdf_normalized[greyscale_image[i, j]]

plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.plot(hist, color='maroon')
plt.title("Original Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.subplot(2, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title("Equalized Image")
plt.axis("off")

equalized_hist = np.zeros(256)
for i in range(rows):
    for j in range(cols):
        equalized_hist[equalized_image[i, j]] += 1

plt.subplot(2, 2, 3)
plt.plot(equalized_hist, color='maroon')
plt.title("Equalized Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
