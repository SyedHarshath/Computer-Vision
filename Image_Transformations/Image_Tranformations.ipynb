```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### IMAGE 1

```python
image = cv2.imread("madara-obito uchiha.jpg")
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

### Image Translation

```python
rows,cols,channels = image_rgb.shape
bx = 1500
by = 900
translated_image = np.zeros_like(image)
for y in range(rows):
    for x in range(cols):
        new_x = x + bx
        new_y = y + by
        if 0<=new_x<cols and 0<=new_y<rows:
            translated_image[new_y,new_x] = image_rgb[y,x]

plt.imshow(translated_image)
plt.axis('off')
plt.title('Translated Image')
plt.show()
```

### Image Rotation

```python
rows,cols,channels = image_rgb.shape
rotated_image = np.zeros_like(image_rgb)
angle = 45
angle_rad = np.radians(angle)
rotated_matrix = np.array([[np.cos(angle_rad),-np.sin(angle_rad)],[np.sin(angle_rad),np.cos(angle_rad)]])
cx , cy = cols//2,rows//2
for y in range(rows):
    for x in range(cols):
        tx,ty = x - cx , y - cy
        new_tx,new_ty = np.dot(rotated_matrix ,np.array([tx,ty]))
        new_x,new_y = int(new_tx + cx),int(new_ty + cy)
        if 0<=new_x<cols and 0<=new_y<rows:
            rotated_image[new_y,new_x] = image_rgb[y,x]

plt.imshow(rotated_image)
plt.axis('off')
plt.title('Rotated Image')
plt.show()
```

### Image Scaling

```python
sx ,sy = 2.5,1.0
scaled_image = np.zeros_like(image_rgb)
cx ,cy = cols//2,rows//2
for y in range(rows):
    for x in range(cols):
        tx , ty = x - cx , y - cy
        new_tx,new_ty = tx * sx,ty * sy
        new_x,new_y = int(new_tx + cx),int(new_ty + cy)

        if 0<=new_x<cols and 0<=new_y<rows:
            scaled_image[new_y,new_x] = image_rgb[y,x]

plt.imshow(scaled_image)
plt.axis('off')
plt.title('Scaled Image')
plt.show()
```

### Image Shearing

```python
shear_x, shear_y = 0.5, 0.5
sheared_image = np.zeros_like(image_rgb)
cx, cy = cols // 2, rows // 2
for y in range(rows):
    for x in range(cols):
        tx, ty = x - cx, y - cy
        new_tx = tx + shear_x * ty
        new_ty = ty + shear_y * tx
        new_x, new_y = int(new_tx + cx), int(new_ty + cy)
        if 0 <= new_x < cols and 0 <= new_y < rows:
            sheared_image[new_y, new_x] = image_rgb[y, x]

plt.imshow(sheared_image)
plt.axis('off')
plt.title('Sheared Image')
plt.show()
```

### Image Reflection

```python
horizontal_reflection = np.zeros_like(image_rgb)
vertical_reflection = np.zeros_like(image_rgb)
diagonal_reflection = np.zeros_like(image_rgb)
for y in range(rows):
    for x in range(cols):
        horizontal_reflection[y, x] = image_rgb[y, cols - x - 1]
        vertical_reflection[y, x] = image_rgb[rows - y - 1, x]
        diagonal_reflection[y, x] = image_rgb[rows - y - 1, cols - x - 1]

plt.figure(figsize=(10, 5))
plt.subplot(1, 4, 1)
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Original Image")

plt.subplot(1, 4, 2)
plt.imshow(horizontal_reflection)
plt.axis('off')
plt.title("Horizontal Reflection")

plt.subplot(1, 4, 3)
plt.imshow(vertical_reflection)
plt.axis('off')
plt.title("Vertical Reflection")

plt.subplot(1, 4, 4)
plt.imshow(diagonal_reflection)
plt.axis('off')
plt.title("Diagonal Reflection")
plt.show()
```

### Affine Transformations

```python
rows,cols,channels = image_rgb.shape
print(rows,cols,channels)
```

```python
def affine_transform(image, matrix):
    rows, cols, channels = image.shape
    output = np.zeros_like(image)
    for y in range(rows):
        for x in range(cols):
            new_coords = np.dot(matrix[:, :2], [x, y]) + matrix[:, 2]
            new_x, new_y = int(new_coords[0]), int(new_coords[1])
            if 0 <= new_x < cols and 0 <= new_y < rows:
                output[y, x] = image[new_y, new_x]
    return output
image = cv2.imread("madara-obito uchiha.jpg")
angle = 30
scale = 1.8
tx, ty = 20, -50
Sx, Sy = 0.3, 0.2
theta = np.radians(angle)
cos_val, sin_val = np.cos(theta), np.sin(theta)
matrix = np.array([
    [cos_val * scale + Sx * sin_val, -sin_val * scale + Sx * cos_val, tx],
    [Sy * cos_val + sin_val * scale, Sy * sin_val + cos_val * scale, ty]
])
transformed_image = affine_transform(image_rgb, matrix)
plt.subplot(1,2,1)
plt.axis('off')
plt.imshow(image_rgb)
plt.title("Original Image")
plt.show()
plt.subplot(1,2,2)
plt.axis('off')
plt.imshow(transformed_image)
plt.title("Transformed Image")
plt.show()
```
