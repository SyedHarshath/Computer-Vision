```python
import gradio as gr
from transformers import pipeline

# Load a pre-trained sentiment analysis model
sentiment_model = pipeline("sentiment-analysis")

# Define the sentiment analysis function
def analyze_sentiment(text):
    result = sentiment_model(text)
    label = result[0]['label']
    score = result[0]['score']
    return f"Sentiment: {label}, Confidence: {score:.2f}"

# Create the Gradio interface
demo = gr.Interface(
    fn=analyze_sentiment,  # Function to call
    inputs=gr.Textbox(placeholder="Enter text for sentiment analysis..."),  # Input component
    outputs="text",  # Output component
    title="Sentiment Analysis App",  # Title of the app
    description="Analyze the sentiment of text as Positive, Negative, or Neutral."  # Description
)

# Launch the app
if __name__ == "__main__":
    demo.launch()
```

```python
import gradio as gr
from transformers import pipeline

# Load the text summarization model
summarizer = pipeline("summarization")

# Define the function for text summarization
def summarize_text(text):
    summary = summarizer(text, max_length=150, min_length=40, do_sample=False)
    return summary[0]["summary_text"]

# Create the Gradio interface
demo = gr.Interface(
    fn=summarize_text,  # Function to call
    inputs=gr.Textbox(lines=10, placeholder="Enter text to summarize...", label="Input Text"),  # Input component
    outputs=gr.Textbox(label="Summarized Text"),  # Output component
    title="Text Summarization App",  # Title of the app
    description="Summarize long text into concise information using a pre-trained model."  # Description
)

# Launch the app
if __name__ == "__main__":
    demo.launch()
```

```python
import gradio as gr
import requests

def check_grammar(text):
    try:
        # Make a POST request to LanguageTool API
        response = requests.post(
            "https://api.languagetool.org/v2/check",  # API endpoint
            data={"text": text, "language": "en-US"},  # Specify English language
            timeout=10
        )
        response.raise_for_status()  # Raise an error for bad responses
        results = response.json()

        if not results.get('matches'):
            return "No errors found! 🎉"

        error_list = ["**Found errors:**"]
        for error in results['matches']:
            error_list.append(
                f"❌ **Error**: {error['message']}\n"
                f"🔍 **Context**: `{error['context']['text']}`\n"
                f"💡 **Suggestions**: {', '.join([r['value'] for r in error['replacements'][:3]])}"
            )

        return "\n\n".join(error_list)

    except Exception as e:
        return f"Error checking grammar: {str(e)}"

# Create the Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("# Grammar & Spell Checker 📝")

    input_text = gr.Textbox(
        label="Enter your text",
        lines=5,
        placeholder="Type your text here..."
    )

    check_btn = gr.Button("Check Text")
    output = gr.Markdown(label="Results")

    check_btn.click(
        fn=check_grammar,
        inputs=input_text,
        outputs=output
    )

# Launch the app
if __name__ == "__main__":
    demo.launch()
```

```python
import gradio as gr
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load your dataset
df = pd.read_csv('/content/ODI data.csv')

# Print column names for verification
print("Columns in dataset:", df.columns.tolist())

# Replace 'Player' or 'Description' based on your choice
df['Description'] = df['Player'] + ' scored ' + df['Runs'].astype(str) + ' runs with a high score of ' + df['HS'].astype(str)
text_column_name = 'Description'
  # Or use 'Description' if combining columns

# Create TF-IDF vectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df[text_column_name])

def retrieve_information(query):
    # Transform the query
    query_vector = vectorizer.transform([query])

    # Calculate cosine similarity
    similarities = cosine_similarity(query_vector, tfidf_matrix).flatten()

    # Get top 5 most similar entries
    top_indices = similarities.argsort()[-5:][::-1]

    results = []
    for idx in top_indices:
        results.append(f"Record {idx}: {df[text_column_name][idx]}...")

    return "\n\n".join(results)

# Create Gradio interface
iface = gr.Interface(
    fn=retrieve_information,
    inputs=gr.Textbox(lines=2, placeholder="Enter your query here..."),
    outputs="text",
    title="Information Retrieval System",
    description="Enter a query to retrieve relevant information from the dataset."
)

# Launch the interface
iface.launch()
```

```python
import gradio as gr
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import nltk
from nltk.corpus import stopwords

# Ensure NLTK stopwords are downloaded
nltk.download('stopwords')
stop_words = stopwords.words('english')

# Sample dataset (replace with your own data)
data = {
    "documents": [
        "Machine learning is a field of artificial intelligence.",
        "Deep learning is a subset of machine learning.",
        "Natural language processing enables computers to understand human language.",
        "Computer vision is used for image recognition tasks.",
        "Reinforcement learning is a type of machine learning focused on decision-making."
    ]
}
df = pd.DataFrame(data)

# Define topic modeling function
def topic_modeling(n_topics):
    # Preprocess text data
    vectorizer = CountVectorizer(stop_words=stop_words)
    doc_term_matrix = vectorizer.fit_transform(df['documents'])

    # Apply LDA for topic modeling
    lda_model = LatentDirichletAllocation(n_components=n_topics, random_state=42)
    lda_model.fit(doc_term_matrix)

    # Extract topics and words
    topics = []
    for idx, topic in enumerate(lda_model.components_):
        words = [vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-10:]]
        topics.append(f"Topic {idx + 1}: {', '.join(words)}")

    return "\n".join(topics)

# Create Gradio interface
iface = gr.Interface(
    fn=topic_modeling,
    inputs=gr.Slider(minimum=2, maximum=5, step=1, label="Number of Topics"),
    outputs="text",
    title="Topic Modeling App",
    description="Enter the number of topics to extract from the dataset."
)

# Launch the app
if __name__ == "__main__":
    iface.launch()
```

```python
import gradio as gr
from transformers import pipeline

# Load a pre-trained relation extraction model (example: Hugging Face pipeline for named entity recognition)
ner_pipeline = pipeline("ner", model="dslim/bert-base-NER")  # Named Entity Recognition
relation_pipeline = pipeline("text-classification", model="Jean-Baptiste/roberta-large-ner-english")  # Relation Classification

def relation_extraction(text):
    # Step 1: Perform Named Entity Recognition (NER)
    entities = ner_pipeline(text)

    # Extract entities and their types
    extracted_entities = []
    for entity in entities:
        extracted_entities.append(f"{entity['word']} ({entity['entity']})")

    # Step 2: Perform Relation Classification (if multiple entities are found)
    if len(extracted_entities) < 2:
        return "Not enough entities detected for relation extraction."

    relation_results = relation_pipeline(text)

    # Format the output
    results = f"**Detected Entities:**\n{', '.join(extracted_entities)}\n\n"
    results += f"**Predicted Relation:**\n{relation_results[0]['label']} (Confidence: {relation_results[0]['score']:.2f})"

    return results

# Create Gradio interface
iface = gr.Interface(
    fn=relation_extraction,
    inputs=gr.Textbox(lines=4, placeholder="Enter text containing relationships between entities..."),
    outputs="text",
    title="Relation Extraction App",
    description="Extract relationships between entities in text using pre-trained models."
)

# Launch the app
if __name__ == "__main__":
    iface.launch()
```
