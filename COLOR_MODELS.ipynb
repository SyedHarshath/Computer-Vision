### IMAGE 1

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("naruto.jpg")
```

```python
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.title('Original Image')
plt.show()
```

```python
rgb = image_rgb/255.0
cmy = 1 - rgb
cmy_image = (cmy*255).astype(np.uint8)
plt.imshow(cmy_image)
plt.axis('off')
plt.title('CMY Image')
plt.show()
```

```python
hsv_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)
plt.imshow(hsv_image)
plt.axis('off')
plt.title('HSV Image')
plt.show()
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 4))
red_channel = image_rgb.copy()
red_channel[:, :, 1] = 0
red_channel[:, :, 2] = 0

plt.subplot(1, 3, 1)
plt.imshow(red_channel)
plt.title('Red Channel')
plt.axis('off')

green_channel = image_rgb.copy()
green_channel[:, :, 0] = 0
green_channel[:, :, 2] = 0

plt.subplot(1, 3, 2)
plt.imshow(green_channel)
plt.title('Green Channel')
plt.axis('off')

blue_channel = image_rgb.copy()
blue_channel[:, :, 0] = 0
blue_channel[:, :, 1] = 0

plt.subplot(1, 3, 3)
plt.imshow(blue_channel)
plt.title('Blue Channel')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

plt.figure(figsize=(12, 4))
hue = image_hsv[:, :, 0]
hue_colormap = cv2.applyColorMap(hue, cv2.COLORMAP_HSV)

plt.subplot(1, 3, 1)
plt.imshow(hue_colormap)
plt.title('Hue Channel')
plt.axis('off')

sat = image_hsv[:, :, 1]


plt.subplot(1, 3, 2)
plt.imshow(sat,cmap='gray')
plt.title('Saturation Channel')
plt.axis('off')

val = image_hsv[:, :, 2]

plt.subplot(1, 3, 3)
plt.imshow(val,cmap='gray')
plt.title('Value Channel')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18, 12))

for i in range(1, 7):

    random_patch = np.random.rand(5, i + 1, 3)
    random_patch_img = (random_patch * 255).astype(np.uint8)

    hue_patch = np.zeros((5, i + 1, 3), dtype=np.uint8)
    hue_values = np.linspace(0, 179, i + 1, dtype=np.uint8)

    for col in range(i + 1):
        hue_patch[:, col] = [hue_values[col], 255, 255]

    hue_patch_img = cv2.cvtColor(hue_patch, cv2.COLOR_HSV2RGB)

    cmy_patch = 1 - random_patch
    cmy_patch_img = (cmy_patch * 255).astype(np.uint8)

    plt.subplot(3, 6, i)
    plt.imshow(random_patch_img)
    plt.axis('off')
    plt.title(f'RGB Random {i + 1} Columns')

    plt.subplot(3, 6, i + 12)
    plt.imshow(cmy_patch_img)
    plt.axis('off')
    plt.title(f'CMY Random {i + 1} Columns')

    plt.subplot(3, 6, i + 6)
    plt.imshow(hue_patch_img)
    plt.axis('off')
    plt.title(f'HSV Hue {i + 1} Columns')

plt.tight_layout()
plt.show()
```

```python
hue_add = (hue + 45) % 180
hue_sub = (hue - 45) % 180
hue_mul = (hue * 14) % 180

hsv_add = image_hsv.copy()
hsv_add[:, :, 0] = hue_add

hsv_sub = image_hsv.copy()
hsv_sub[:, :, 0] = hue_sub

rgb_add = cv2.cvtColor(hsv_add, cv2.COLOR_HSV2RGB)
rgb_sub = cv2.cvtColor(hsv_sub, cv2.COLOR_HSV2RGB)

hsv_mul = image_hsv.copy()
hsv_mul[:, :, 0] = hue_mul

rgb_add = cv2.cvtColor(hsv_add, cv2.COLOR_HSV2RGB)
rgb_sub = cv2.cvtColor(hsv_sub, cv2.COLOR_HSV2RGB)
rgb_mul = cv2.cvtColor(hsv_mul, cv2.COLOR_HSV2RGB)

plt.subplot(2, 3, 1)
plt.imshow(rgb_add)
plt.title('Hue + 45 (RGB)')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(rgb_sub)
plt.title('Hue - 45 (RGB)')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(rgb_mul)
plt.title('Hue * 14 (RGB)')
plt.axis('off')

original_rgb = cv2.cvtColor(image_hsv, cv2.COLOR_HSV2RGB)
plt.subplot(2, 3, 5)
plt.imshow(original_rgb)
plt.title('Original (RGB)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
